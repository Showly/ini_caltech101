```python
import json
def plot(history_json):
    with open(history_json) as data_file:    
        data = json.load(data_file)
    
    fig, axs = plt.subplots(1,2)
    pd.DataFrame(data['loss']).plot(ax=axs[0])
    pd.DataFrame(data['val_loss']).plot(ax=axs[0])

    pd.DataFrame(data['acc']).plot(ax=axs[1], ylim=(0,1))
    pd.DataFrame(data['val_acc']).plot(ax=axs[1], ylim=(0,1))
    return data
```
