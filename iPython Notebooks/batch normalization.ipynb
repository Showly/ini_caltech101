```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import pprint
%matplotlib inline
```

```python
import json
def plot(history_json):
    with open(history_json) as data_file:    
        data = json.load(data_file)
    
    fig, axs = plt.subplots(1,2)
    pd.DataFrame(data['loss']).plot(ax=axs[0], figsize=(20,10), title='loss')
    pd.DataFrame(data['val_loss']).plot(ax=axs[0])

    pd.DataFrame(data['acc']).plot(ax=axs[1], ylim=(0,1), figsize=(20,10), title='accuracy')
    pd.DataFrame(data['val_acc']).plot(ax=axs[1], ylim=(0,1))
    return data
```

```python
def show_arch(architecture_json):
    with open(architecture_json) as data_file:
        data = json.load(data_file)
    pprint.pprint(data)
```
