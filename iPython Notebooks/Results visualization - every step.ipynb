```python
import json
def plot(history_json):
    with open(history_json) as data_file:    
        data = json.load(data_file)
    
    fig, axs = plt.subplots(2,2)
    
    df_loss = pd.DataFrame(np.array(data['loss']).flatten(), columns=['train loss'])
    df_loss.plot(ax=axs[0][0], figsize=(20,20), title='train loss')

    df_acc = pd.DataFrame(np.array(data['acc']).flatten(), columns=['train acc'])
    df_acc.plot(ax=axs[0][1], ylim=(0,1), title='train acc')
    
    mean_loss = np.mean(data['loss'], axis=1)
    mean_val_loss = np.mean(data['val_loss'], axis=1)
    mean_acc = np.mean(data['acc'], axis=1)
    mean_val_acc = np.mean(data['val_acc'], axis=1)
    
    pd.DataFrame(mean_loss, columns=['train loss']).plot(ax=axs[1][0], xlim=(0, len(mean_loss)), title='loss mean per epoch')
    pd.DataFrame(mean_val_loss, columns=['val loss']).plot(ax=axs[1][0])

    pd.DataFrame(mean_acc, columns=['train acc']).plot(ax=axs[1][1], ylim=(0,1), xlim=(0, len(mean_acc)), title='acc mean per epoch')
    pd.DataFrame(mean_val_acc, columns=['val acc']).plot(ax=axs[1][1], ylim=(0,1))
    
    print "val acc:", mean_val_acc
    return data
```

```python
def show_arch(architecture_json):
    with open(architecture_json) as data_file:
        data = json.load(data_file)
    pprint.pprint(data)
```
